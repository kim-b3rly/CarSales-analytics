### Display the number of cars per fuel system using a bar chart

```python
import matplotlib.pyplot as plt

#extracting data for the y axis and x axis

#y axis will be the count for each fuel system
fsystem_count=car_df['fuelsystem'].value_counts()

#x axis will be each fuel system
fuel_system=car_df['fuelsystem'].unique()

#plotting the bar chart

fig=plt.figure()

plt.bar(fuel_system,fsystem_count,color='purple')

plt.xlabel("Types of fuel system")
plt.ylabel("Number of cars")

plt.title("Number of cars per fuel system")

plt.show() # show the graph
```

### Display visualisation to present customer behaviour

```python
#Creating a pie chart with car body to get insight on lifestyle 

#find the count per car body and print to screen
carbody_counts=car_df['carbody'].value_counts()

#calculate percentages of each count

percents=[]
for i in range(len(carbody_counts)):
    percents.append(carbody_counts[i]/sum(carbody_counts))

#store car body names in a list

body_names=["sedan","hatchback","hardtop","convertible"]
```

```python
#find the unique car body values

names=["sedan","hatchback","wagon","hardtop","convertible"]
```

```python
import matplotlib.pyplot as plt

#create an object for plotting 

fig=plt.figure(figsize=(10,8))

#plotting a pie chart with a title and a legend

plt.pie(percents,labels=names)

plt.title("Distribution of car body types in sales")

plt.legend(loc='upper left')

plt.show()
```
